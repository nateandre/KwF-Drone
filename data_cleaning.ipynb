## Data cleaning and augmentation

This is meant as an example for cleaning the data given the implicit structure laid out in the data labeling guide. As of now, there is an assumption that images collected are of shape 722x398x3 (as this is the size of the image used in this notebook) - this can be relaxed later. This notebook is being used to augment a single image to be used by the YOLO3 inspired model. This image needs to be adjusted to a size of 608x608 in order to create valid input that will be reduced to a 19x19x29 grid. Images in this example will be augmented by horizontal reflection.

### Data cleaning

```python
image_path = "../data/kwf/raw/" # directory which holds all of the images
labels_path = "../data/kwf/raw/" # directory which holds all of the labels for images
```

```python
# image parameters
image_height = 398
image_width = 722

new_height = 608
new_width = 608
```

```python
"""
Does horizontal reflection for each image and updates encodings and returns both 
"""
def double_data(img,lis_info):
    img = np.flip(img, axis=1)
    lis_info_len = len(lis_info)
    lis_info_new = []
    
    for i in range(0,lis_info_len):
        xmin = lis_info[i][1]
        width = lis_info[i][3]
        xmin = new_width - xmin - width
        
        lis_info_new.append([lis_info[i][0],xmin,lis_info[i][2],lis_info[i][3],lis_info[i][4]])
    
    return img,lis_info_new
```

```python
"""
Updates the bounding box coordinates such that they work in (680,680) resolution after interpolation
Returns the updated encoding for a given image
"""
def update_bounding_box(lis_info):
    lis_info_len = len(lis_info)
    scale_x = new_width / image_width # determining how each coordinate should be scaled
    scale_y = new_height / image_height
    
    for i in range(0,lis_info_len):
        xmin = lis_info[i][1]
        width = lis_info[i][3]
        xmax = xmin + width
        
        ymin = lis_info[i][2]
        height = lis_info[i][4]
        ymax = ymin + height
        
        xmin = xmin * scale_x # updating the values for x-coordinates
        xmax = xmax * scale_x
        width = xmax - xmin
        
        ymin = ymin * scale_y # updating the values for y-coordinates
        ymax = ymax * scale_y
        height = ymax - ymin
        
        lis_info[i] = [lis_info[i][0],xmin,ymin,width,height]

    return lis_info
```

```python
"""
Scales the image to (new_width, new_height)
Returns image of shape (new_width,new_height,3)
"""
def transform_image(path,file):
    img = cv2.imread(path+file)
    img = cv2.resize(img, (new_width,new_height), interpolation = cv2.INTER_AREA) # scaled
    img = np.array(img)
    img.shape = (new_width,new_height,3)
    return img
```

```python
# cleans the information for each image by removing the name and changing the coordinates to integers
# changes the coordinate types from string to integer, and also updates the coordinates of the bounding boxes
# for the image based on the new image shape (680,680,3)
def clean_lis_info(lis_info):
    lis_info_len = len(lis_info)
    for i in range(0,lis_info_len):
        lis_info[i] = [lis_info[i][1]] + [int(lis_info[i][x]) for x in range(2,6)]
    
    lis_info = update_bounding_box(lis_info)
    return lis_info
```

```python
# takes in a list of information an image - [name,type,x,y,width,height]
# returns a tuple which includes the encoding for the image and the image itself
def get_img_encoding(lis_info,image_path):
    name = lis_info[0][0]
    lis_info = clean_lis_info(lis_info)
    img = transform_image(image_path,name) # getting the image reshaped to (680,680,3)
    img2,lis_info2 = double_data(img,lis_info)
    return [(img,lis_info),(img2,lis_info2)]
```

```python
# Takes in a labels file and returns the encoding for each of the images within the file
# returns a list of encodings for each of the images along with the image's pixels, as tuples
def get_encoding_data(image_path,labels_path,file):
    encodings = []
    with open(labels_path+file) as afile: # opening a labels file
        lines = afile.readlines()
        lis_info  = [] # holds info temporarily for each image
        old_name = lines[0].split(" ")[0] # first image file name
        
        for line in lines:
            split_line = line.strip("\n").split(" ")
            name = split_line[0] # name of image - unique for each image
            if (name != old_name): # found a new image file
                encoding = get_img_encoding(lis_info,image_path) # getting encoding for given info for a single image
                encodings += encoding
                lis_info = [] # resetting info holder for new image
                
            lis_info.append(split_line)
            old_name = name
            
        encoding = get_img_encoding(lis_info,image_path) # getting encoding for last image in file
        encodings += encoding
        
        return encodings
```

```python
# Gets the YOLO encodings for labeled images
# returns a list of tuples - (img,2D array of encodings)
def get_all_data(image_path,labels_path):
    all_encodings = []
    for file in os.listdir(labels_path):
        if "turtle_boxes" in file: # valid labels file
            img_encodings = get_encoding_data(image_path,labels_path,file)
            all_encodings += img_encodings
            
    return all_encodings
```

```python
# Getting the actual images with associated encodings
example_data = get_all_data(image_path,labels_path)
```

### Creating y-training set based on image encodings

```python
pixels_per_cell = int(new_width / 19) # we want 19 cells
print("pixels per cell:",pixels_per_cell)
```

```python
# returns a list corresponding to the object name being represented as a categorical variable
def get_obj_enc(enc_num):
    enc_lis = [0 for i in range(0,24)] # initialize
    enc_lis[enc_num] = 1
    return enc_lis
```

```python
encoded_data = []
image_data = []

for i in range(len(example_data)):
    encode = np.zeros((19,19,29))
    encoding_data = example_data[i][1]
    for enc in encoding_data: # going through each bounding box per image
        name = enc[0] # name of the object in bounding box
        x1 = enc[1] # reading in the bounding box data 
        y1 = enc[2]
        width = enc[3]
        height = enc[4]
        # Getting the midpoints for this bounding box
        x_mid = x1 + (1/2*width)
        y_mid = y1 + (1/2*height)
        # determine which cell the midpoint is in
        cell_x = x_mid // pixels_per_cell
        cell_y = y_mid // pixels_per_cell
        # update x_mid and y_mid to be in relation of cell
        x_mid = (x_mid - (pixels_per_cell * cell_x)) / pixels_per_cell
        y_mid = (y_mid - (pixels_per_cell * cell_y)) / pixels_per_cell
        # change the height and width to be divided by the image height and width
        width = width / new_width
        height = height / new_height
        # create encoding for the object corresponding to the bounding box
        obj_enc_num = encoding_dict[name] - 1 
        obj_enc = get_obj_enc(obj_enc_num)
        # entry for a specific cell within the y-data-representation
        entry = np.array([1,x_mid,y_mid,width,height]+obj_enc)
        # adding bounding box encoding to the y-data-representation
        encode[int(cell_x),int(cell_y),:] = entry
        
    encoded_data.append(encode)
    image_data.append(example_data[i][0])
```

```python
# Printing an example of an encoding
img = image_data[1]
encoding = encoded_data[1]
boxes = []
cells = np.nonzero(encoding[:,:,0]) # Get the index tuple of the encoding
for i in range(0,len(cells[0])): # looping through all objects in image
    the_encoding = encoding[cells[0][i],cells[1][i],:]
    x_mid = the_encoding[1]
    y_mid = the_encoding[2]
    width = the_encoding[3]
    height = the_encoding[4]
    width = width * new_width
    height = height * new_height
    
    x_mid = x_mid * pixels_per_cell + (pixels_per_cell * cells[0][i])
    y_mid = y_mid * pixels_per_cell + (pixels_per_cell * cells[1][i])
    x_min = x_mid - (1/2*width)
    y_min = y_mid - (1/2*height)
    boxes.append([x_mid,y_mid,x_min,y_min,width,height]) # data for each of the bounding boxes
```

```python
# printing out how the y-encodings relate to the bounding boxes over the initial image
x_ticks = np.arange(0, new_width, pixels_per_cell)
y_ticks = np.arange(0, new_width, pixels_per_cell)
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)
ax.grid(which='both',color='r')
ax.imshow(img)
for box in boxes: # adding all of the bounding boxes from y-encoding
    bounding_box = patches.Rectangle((box[2],box[3]),box[4],box[5],linewidth=1,edgecolor='b',facecolor='none')
    midpoint = patches.Circle((box[0],box[1]), radius=10)
    ax.add_patch(bounding_box)
    ax.add_patch(midpoint)
plt.show()
```

```python
# normalizing all images by dividing by 255
for i in range(len(image_data)):
    image_data[i] = image_data[i] / 255
```

```python
# creating the final input for the yolo model
images = np.array(image_data)
encodings = np.array(encoded_data)
print(images.shape)
print(encodings.shape)
```
